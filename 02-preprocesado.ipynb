```python
"""
Luego de mostrar los datos, se va a realizar una  limpieza y preprocesado de los mismos para mirar que tan influyentes pueden ser para el modelo
que se va a construir para predecir el rendimiento global de los estudiantes.En este caso, existe la posibilidad de convertir varias columnas que estan en forma categorica
a variables numericas, por ejemplo las horas a la semana que trabaja cada estudiante,el estrato de vivienda, pago matricula propia, etc.En concreto se va a realizar para dos:el
numero de horas a la semana que trabaja cada estudiante y el estrato de vivienda. Vamos a realizar un one-hot encoding para las dos columnas.El numero de horas
a la semana que trabaja la vamos a convertir en un one-hot encoding y el estrato de vivienda en uno continuo.
"""
y= y[['ESTU_HORASSEMANATRABAJA', 'FAMI_ESTRATOVIVIENDA', 'RENDIMIENTO_GLOBAL']]




```

```python
#Usamos la funcion isna, para ver cuantos son los valores perdidos en estas columnas
sum(y.ESTU_HORASSEMANATRABAJA.isna()), sum(y.FAMI_ESTRATOVIVIENDA.isna())
```

```python
#Por lo tanto, se van a sustituir los valores perdidos con el simbolo de que no hay informacion para las columnas de estrato de vivienda y las horas a la semana que trabaja cada estudiante

y.ESTU_HORASSEMANATRABAJA.values[y.ESTU_HORASSEMANATRABAJA.isna()] = 'no info'
y.FAMI_ESTRATOVIVIENDA.values[y.FAMI_ESTRATOVIVIENDA.isna()] = 'no info'
sum(y.ESTU_HORASSEMANATRABAJA.isna()), sum(y.FAMI_ESTRATOVIVIENDA.isna())
```

```python
#se convierten en variables numericas el valor de la columna del estrato de vivienda de la familia
import numpy as np
#se crea una copia del dataframe
df=y.copy()


#se le asignan valores númericos a cada uno de los estratos y a los datos faltantes de la columna del estrato de vivienda
cmap={'Estrato 6':6,
      'Estrato 5':5,
      'Estrato 4':4,
      'Estrato 3':3,
      'Estrato 2':2,
      'Estrato 1':1,
      'sin estrato':0,
      'no info':-1}
y.FAMI_ESTRATOVIVIENDA=df.FAMI_ESTRATOVIVIENDA.map(cmap)
y.FAMI_ESTRATOVIVIENDA.unique()
y.FAMI_ESTRATOVIVIENDA.value_counts()


#Se observan los estratos en los que viven cada uno de los estudiantes con sus respectivos valores númericos






















```

```python
#se limpia la columna de las horas a la semana que trabaja cada estudiante
df.ESTU_HORASSEMANATRABAJA=['no info' if i in ['No sabe', 'No Aplica'] else i for i in df.ESTU_HORASSEMANATRABAJA.values]
df.ESTU_HORASSEMANATRABAJA.unique()
```

```python
#se muestran los datos de las horas a la semana que trabajan cada uno de los estudiantes y tambíen de aquellos que no se tiene información.
df.ESTU_HORASSEMANATRABAJA.value_counts()
```

```python
#Luego, se va a crear un one-hot encoding , primero se va a hacer un mapeo
sorted_series=sorted
sorted_series
x=y.ESTU_HORASSEMANATRABAJA.sort_values()
x
ESTU_HORASSEMANATRABAJA_vals=x.unique()
ESTU_HORASSEMANATRABAJA_vals
ESTU_HORASSEMANATRABAJA_vals_onehot_vals={val: np.eye(len(ESTU_HORASSEMANATRABAJA_vals))[i] for i,val in enumerate(ESTU_HORASSEMANATRABAJA_vals)}
ESTU_HORASSEMANATRABAJA_vals_onehot_vals
```

```python
ESTU_HORASSEMANATRABAJA_onehot_enc = np.r_[[ESTU_HORASSEMANATRABAJA_vals_onehot_vals[i] for i in y.ESTU_HORASSEMANATRABAJA]]
ESTU_HORASSEMANATRABAJA_onehot_enc
```

```python
#Posteriormente, se aplica el one-hot encoding y todas las columnas quedan en un solo dataframe
ESTU_HORASSEMANATRABAJA_df = pd.DataFrame(ESTU_HORASSEMANATRABAJA_onehot_enc, columns=[f"ESTU_HORASSEMANATRABAJA__{v}" for v in ESTU_HORASSEMANATRABAJA_vals_onehot_vals])
ESTU_HORASSEMANATRABAJA_df
```

```python
#Ahora, podemos eliminar la columna de las horas a la semana que trabaja cada estudiante
y = pd.concat([ESTU_HORASSEMANATRABAJA_df, y], axis=1).drop('ESTU_HORASSEMANATRABAJA', axis=1)
y.shape
```

```python
"""
Ahora, se tiene un nuevo dataframe, con las columnas de estrato de vivienda y las horas a la semana que trabaja cada estudiante con sus respectivos valores númericos.Tambien se
cuentan con la columna de rendimiento global de los estudiantes. Además, se cambian la cantidad de columnas que tiene el dataframe
"""
y.head()
```

```python
#Luego, para poder llevar a este dataset a que todo en el sea numerico, podemos hacer el mismo procedimiento para la columna de rendimiento global
R_col = 'RENDIMIENTO_GLOBAL'

rmap = {'alto': 3, 'bajo':0, 'medio-bajo':1, 'medio-alto':2}
y[R_col] = [rmap[i] for i in y[R_col]]
y.head()

#Finalmente, estos datos nos muestran la columna de las horas a la semana que trabaja cada estudiante con el one-hot enconding y el resto de columnas ya estan en datos numericos.
```
