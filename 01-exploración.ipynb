```python
#se cargan los datos a través de la libreria pandas y se usa matplolib para gráficar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#se carga el train.csv
y=pd.read_csv("train.csv")
print("shape of the dataframe",y.shape)
```

```python
"""se evidencian los datos que están en el train.csv de kaggle,
en el cual se puede apreciar la descripción del estudiante, junto con
las etiquetas de clasificación de desempeño de rendimiento global y
el Id de cada estudiante"""
y.head()
```

```python

"""
Se ilustran tres  columnas del dataframe, las cuales corresponden al
estrato de la vivienda, el programa de departamento y el rendimiento global
"""
z=y[["FAMI_ESTRATOVIVIENDA","ESTU_PRGM_DEPARTAMENTO","RENDIMIENTO_GLOBAL"]]
z.head()
```

```python
#Se realiza la exploración de los datos
#Se muestra una gráfica de barras con  la distribución de la cantidad de estudiantes que pertenecen a cada uno de los estratos.
z.FAMI_ESTRATOVIVIENDA.value_counts().plot(kind="bar", color="GREEN")
plt.xlabel("Estrato")
plt.ylabel("Cantidad de estudiantes")
plt.title("Distribución de estudiantes por estrato")
plt.legend()
#En la gráfica se observa que la mayoría de estudiantes pertenecen al estrato 2.
```

```python
#se muestran los estudiantes que hay por programa en cada uno de los departamentos
z.ESTU_PRGM_DEPARTAMENTO.value_counts().plot(kind="bar",color="red")
plt.xlabel("Departamento")
plt.ylabel("Cantidad de estudiantes")
plt.title("Distribución de estudiantes por departamento")
"""
En la gráfica se muestra, el número de estudiantes que hay por cada uno de los
departamentos, en la cual se observa que la mayoría de estudiantes  pertenecen a Bógota
y Antioquia
"""
```

```python
from rlxutils import subplots
import matplotlib.pyplot as plt
"""
se analiza cual fue la tendencia del rendimiento global en cada
uno de los departamentos.
En la gráfica se puede evidenciar  que en Bogota y Antioquia, la mayoría
de estudiantes obtuvo un rendimiento alto
"""
c = sorted(z.ESTU_PRGM_DEPARTAMENTO.value_counts().index)
for ax,ci in subplots(c, n_cols=4, usizex=4):
    zc = z[z.ESTU_PRGM_DEPARTAMENTO==ci]
    counts=zc.RENDIMIENTO_GLOBAL.value_counts()
    filtered_counts=counts[counts.index.isin(['bajo', 'medio-bajo', 'medio-alto', 'alto'])]
    filtered_counts.plot(kind='bar',color="Red")
    plt.title(ci)
    plt.xlabel("RENDIMIENTO_GLOBAL")
    plt.ylabel("Cantidad de estudiantes")

```

```python
#se pueden calcular algunas variables estadisticas con los datos
#Por ejemplo, se puede hallar el promedio  de estudiantes que hay en total en base al programa por departamento
z.groupby("ESTU_PRGM_DEPARTAMENTO").ESTU_PRGM_DEPARTAMENTO.count()
promedio=z.groupby("ESTU_PRGM_DEPARTAMENTO").ESTU_PRGM_DEPARTAMENTO.count().mean()
print(promedio)
#En promedio hay 22338.709 estudiantes en base a su programa por departamento
```

```python
# También, se calcula cuantos estudiantes en promedio de Antioquia, asistieron a la prueba
z.groupby("ESTU_PRGM_DEPARTAMENTO").ESTU_PRGM_DEPARTAMENTO.count()["ANTIOQUIA"]
promedio=z.groupby("ESTU_PRGM_DEPARTAMENTO").ESTU_PRGM_DEPARTAMENTO.count()["ANTIOQUIA"].mean()
print(promedio)
#En promedio en Antioquia, hay 83607.0 estudiantes que realizaron la prueba
```

```python
"""
Luego se compara cual fue la tendencia del rendimiento global de acuerdo
a cada uno de los estratos de la vivienda.Se podría decir entonces, que estos
pueden influenciar en el rendimiento. Por ejemplo  de acuerdo a las gráficas,
se observa que al comparar el estrato 1 con el 6, en el primero predomina
un rendimiento bajo y en el segundo un rendimiento alto
"""
c = sorted(z.FAMI_ESTRATOVIVIENDA.value_counts().index)
for ax,ci in subplots(c, n_cols=4, usizex=4):
    zc = z[z.FAMI_ESTRATOVIVIENDA==ci]
    counts=zc.RENDIMIENTO_GLOBAL.value_counts()
    filtered_counts=counts[counts.index.isin(['bajo', 'medio-bajo', 'medio-alto', 'alto'])]
    filtered_counts.plot(kind='bar',color="GREEN")
    plt.title(ci)
    plt.xlabel("RENDIMIENTO_GLOBAL")
    plt.ylabel("Cantidad de estudiantes")
```

```python
#Posteriormente, se calcula  la desviación estándar para la columna de los estratos
desviacion=z.groupby("FAMI_ESTRATOVIVIENDA").FAMI_ESTRATOVIVIENDA.count().std()
print(desviacion)
```

```python
#También se puede calcular cual es la desviación por cado uno de los estratos
#En este caso se calcula  la desviación estándar del estrato 1
desviacion=z.groupby("FAMI_ESTRATOVIVIENDA").FAMI_ESTRATOVIVIENDA.count()["Estrato 1"].std()
print(desviacion)
```
