```python
"""
Luego de mostrar los datos, se va a realizar un  preprocesado de los mismos para mirar que tan influyentes pueden ser para el modelo
que se va a construir para predecir el rendimiento global de los estudiantes.En este caso, existe la posibilidad de convertir varias columnas que estan en forma categorica
a variables numericas, por ejemplo las horas a la semana que trabaja cada estudiante,el estrato de vivienda, pago matricula propia, etc.En concreto se va a realizar para las
columnas de número de horas a la semana que trabaja cada estudiante,  el estrato de vivienda, el valor de la matricula de la universidad,fami_tieneinternet,
educación padre, la columna de si el estudiante paga matricula propia .
 Vamos a realizar un one-hot encoding para la columna de el numero de horas
a la semana que trabaja cada estudiante y la educacion del padre y la madre y vamos a convertir la columna de el estrato de vivienda y valor de la
matricula de la universidad y la columna de si el estudiante paga matricula propia  en un hot-hot encoding continuo.
"""
Y=pd.read_csv('/content/udea-ai4eng-20242/train.csv')
y=Y.copy()
y= y[['ESTU_HORASSEMANATRABAJA', 'FAMI_ESTRATOVIVIENDA','ESTU_PAGOMATRICULAPROPIO','ESTU_VALORMATRICULAUNIVERSIDAD', 'FAMI_TIENEINTERNET','FAMI_EDUCACIONPADRE', 'RENDIMIENTO_GLOBAL']]



```

```python
#Se observa que en la columna de las horas a la semana que trabaja cada estudiante hay 30857 valores perdidos y en la columna Fami_Estrato vivienda hay 32137
```

```python
#Luego,se sustituyen los valores perdidos faltantes de estas columnas con el simbolo no info

y.ESTU_HORASSEMANATRABAJA.values[y.ESTU_HORASSEMANATRABAJA.isna()] = 'no info'
y.FAMI_ESTRATOVIVIENDA.values[y.FAMI_ESTRATOVIVIENDA.isna()] = 'no info'
y.ESTU_PAGOMATRICULAPROPIO.values[y.ESTU_PAGOMATRICULAPROPIO.isna()] = 'no info'
y.ESTU_VALORMATRICULAUNIVERSIDAD.values[y.ESTU_VALORMATRICULAUNIVERSIDAD.isna()] = 'no info'

y.FAMI_EDUCACIONPADRE.values[y.FAMI_EDUCACIONPADRE.isna()] = 'no info'

sum(y.ESTU_HORASSEMANATRABAJA.isna()), sum(y.FAMI_ESTRATOVIVIENDA.isna()), sum(y.ESTU_PAGOMATRICULAPROPIO.isna()), sum(y.ESTU_VALORMATRICULAUNIVERSIDAD.isna()),sum(y.FAMI_EDUCACIONPADRE.isna())
```

```python
#se convierten en variables numericas el valor de la columna del estrato de vivienda de la familia
import numpy as np
#se crea una copia del dataframe
df=y.copy()


#se le asignan valores númericos a cada uno de los estratos y a los datos faltantes de la columna del estrato de vivienda
cmap={'Estrato 6':6,
      'Estrato 5':5,
      'Estrato 4':4,
      'Estrato 3':3,
      'Estrato 2':2,
      'Estrato 1':1,
      'sin estrato':0,
      'no info':-1}
y.FAMI_ESTRATOVIVIENDA=df.FAMI_ESTRATOVIVIENDA.map(cmap)
y.FAMI_ESTRATOVIVIENDA.unique()
y.FAMI_ESTRATOVIVIENDA.value_counts()


#Se observan los estratos en los que viven cada uno de los estudiantes con sus respectivos valores númericos






















```

```python
#se juntan la informacion de no sabe o no se aplica para las horas a la semana que trabaja cada estudiante
df.ESTU_HORASSEMANATRABAJA=['no info' if i in ['No sabe', 'No Aplica'] else i for i in df.ESTU_HORASSEMANATRABAJA.values]
df.ESTU_HORASSEMANATRABAJA.unique()
```

```python
#se muestran los datos de las horas a la semana que trabajan cada uno de los estudiantes y tambíen de aquellos que no se tiene información.
df.ESTU_HORASSEMANATRABAJA.value_counts()
```

```python
#Luego, se va a crear un one-hot encoding , primero se va a hacer un mapeo
sorted_series=sorted
sorted_series
x=y.ESTU_HORASSEMANATRABAJA.sort_values()
x
ESTU_HORASSEMANATRABAJA_vals=x.unique()
ESTU_HORASSEMANATRABAJA_vals
ESTU_HORASSEMANATRABAJA_vals_onehot_vals={val: np.eye(len(ESTU_HORASSEMANATRABAJA_vals))[i] for i,val in enumerate(ESTU_HORASSEMANATRABAJA_vals)}
ESTU_HORASSEMANATRABAJA_vals_onehot_vals
```

```python
ESTU_HORASSEMANATRABAJA_onehot_enc = np.r_[[ESTU_HORASSEMANATRABAJA_vals_onehot_vals[i] for i in y.ESTU_HORASSEMANATRABAJA]]
ESTU_HORASSEMANATRABAJA_onehot_enc
```

```python
#Posteriormente, se aplica el one-hot encoding y todas las columnas quedan en un solo dataframe
ESTU_HORASSEMANATRABAJA_df = pd.DataFrame(ESTU_HORASSEMANATRABAJA_onehot_enc, columns=[f"ESTU_HORASSEMANATRABAJA__{v}" for v in ESTU_HORASSEMANATRABAJA_vals_onehot_vals])
ESTU_HORASSEMANATRABAJA_df
```

```python
#Luego tomamos la columna del valor de la matricula de la universidad y reemplazamos los valores de cmap
cmap = {'Entre 1 millón y menos de 2.5 millones': .175,
 'Entre 2.5 millones y menos de 4 millones': .325,
 'Menos de 500 mil': .250,
 'Entre 500 mil y menos de 1 millón': .75,
 'Entre 4 millones y menos de 5.5 millones': .475,
 'Más de 7 millones': .775,
 'Entre 5.5 millones y menos de 7 millones': .625,
 'No pagó matrícula': 0,
 'no info': -1}

y.ESTU_VALORMATRICULAUNIVERSIDAD=df.ESTU_VALORMATRICULAUNIVERSIDAD.map(cmap)
y.ESTU_VALORMATRICULAUNIVERSIDAD.unique().sort()
y.ESTU_VALORMATRICULAUNIVERSIDAD.value_counts()
#Posteriormente, la columna queda en datos nnúmericos
```

```python
#luego miramos los valores de no sabe y no aplica en la columna de educacion del padre
y.FAMI_EDUCACIONPADRE = ['no info' if i in ['No sabe', 'No Aplica'] else i for i in y.FAMI_EDUCACIONPADRE.values]
y.FAMI_EDUCACIONPADRE.value_counts() #se muestran los valores de como es la educación del padre
```

```python
#Posteriormente, le aplicamos un one-hot encoding a esta columna
FAMI_EDUCACIONPADRE_vals=y.FAMI_EDUCACIONPADRE.sort_values().unique()
FAMI_EDUCACIONPADRE_vals_onehot_vals={val: np.eye(len(FAMI_EDUCACIONPADRE_vals))[i] for i,val in enumerate(FAMI_EDUCACIONPADRE_vals)}
FAMI_EDUCACIONPADRE_onehot_enc = np.r_[[FAMI_EDUCACIONPADRE_vals_onehot_vals[i] for i in y.FAMI_EDUCACIONPADRE]]
FAMI_EDUCACIONPADRE_df = pd.DataFrame(FAMI_EDUCACIONPADRE_onehot_enc, columns=[f"FAMI_EDUCACIONPADRE__{v}" for v in FAMI_EDUCACIONPADRE_vals_onehot_vals])
FAMI_EDUCACIONPADRE_df
```

```python
#se muestran los valores de los estudiantes que pagan matricula propia y los que no y tambien de los cuales no se tiene informacion

y.ESTU_PAGOMATRICULAPROPIO.value_counts()
```

```python
#se muestran los valores unicos de esta columna
np.unique(y.ESTU_PAGOMATRICULAPROPIO)


```

```python
#se va a asignar valores de 0 para los estudiantes que no pagan matricula propia y valores de 1 para los que si
y.ESTU_PAGOMATRICULAPROPIO.values[y.ESTU_PAGOMATRICULAPROPIO == "Si"] = 1
y.ESTU_PAGOMATRICULAPROPIO.values[y.ESTU_PAGOMATRICULAPROPIO == "No"] = 0
```

```python
#aplicamos lo mismo para la columna de si la familia tiene internet
y.FAMI_TIENEINTERNET.values[y.FAMI_TIENEINTERNET.isna()] = "No"
sum(y.FAMI_TIENEINTERNET.isna())
x = y.FAMI_TIENEINTERNET.values
FAMI_TIENEINTERNET_vals = sorted(np.unique(x))
FAMI_TIENEINTERNET_onehot_vals = {val: np.eye(len(FAMI_TIENEINTERNET_vals))[i] for i,val in enumerate(FAMI_TIENEINTERNET_vals)}
FAMI_TIENEINTERNET_onehot_vals
FAMI_TIENEINTERNET_onehot_enc = np.r_[[FAMI_TIENEINTERNET_onehot_vals[i] for i in y.FAMI_TIENEINTERNET]]
FAMI_TIENEINTERNET_onehot_enc




```

```python
#se muestra la columna anterior con el one-hot
FAMI_TIENEINTERNET_df = pd.DataFrame(FAMI_TIENEINTERNET_onehot_enc, columns=[f"FAMI_TIENEINTERNET__{v}" for v in FAMI_TIENEINTERNET_onehot_vals])
FAMI_TIENEINTERNET_df


```

```python
#concatenamos la columna de Fami tiene internet
y = pd.concat([FAMI_TIENEINTERNET_df, y], axis=1).drop('FAMI_TIENEINTERNET', axis=1)
y.shape
```

```python
#Ahora, podemos concatenar la columna de las horas a la semana que trabaja cada estudiante
y = pd.concat([ESTU_HORASSEMANATRABAJA_df, y], axis=1).drop('ESTU_HORASSEMANATRABAJA', axis=1)
y.shape
```

```python
#hacemos lo mismo para la columna de educacion padre
y = pd.concat([FAMI_EDUCACIONPADRE_df, y], axis=1).drop('FAMI_EDUCACIONPADRE', axis=1)
y.shape
```
